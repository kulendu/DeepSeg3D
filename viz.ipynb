```python
import matplotlib.pyplot as plt
import os
import cv2
import numpy as np
```

```python
DATA_PATH = '/home/kulendu/college_materials/7th sem project/archive/image_data'
CLASSES = ['tumor', 'not_tumor']

for category in CLASSES:
    path = os.path.join(DATA_PATH, category)
    for image in os.listdir(path):
        img_arr = cv2.imread(os.path.join(path, image))
        resized_img = cv2.resize(img_arr, (224, 224))
        plt.imshow(resized_img)
        
        print(f'Default image dimension ---- {img_arr.shape}')
        print(f'Resized image dimension ---- {resized_img.shape}')

        break
    break
```
